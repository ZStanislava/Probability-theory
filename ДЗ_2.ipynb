Задание 1

Устройство, состоящее из пяти независимо работающих элементов, включается за время Т. Вероятность отказа каждого из них за это время равна 0,4. Найти вероятность того, что откажут три элемента. Ответ округлите до четырёх знаков после запятой. В качестве десятичного разделителя используйте точку.

```python
import math

# Given values
n = 5        # total elements
k = 3        # elements that fail
p = 0.4      # probability of failure
q = 1 - p    # probability of success

# Binomial coefficient C_n^k
C_n_k = math.comb(n, k)

# Calculate the probability P(X = k)
P_X_k = C_n_k * (p ** k) * (q ** (n - k))

# Round the result to four decimal places
round(P_X_k, 4)
```

Задание 3

Производится 8 выстрелов по цели, в каждом из которых вероятность попадания равна 0,3. Найти вероятность того, что цель будет поражена хотя бы два раза. Ответ округлите до тысячных. В качестве десятичного разделителя используйте точку.


Для нахождения вероятности хотя бы двух попаданий, проще всего воспользоваться противоположным событием — вероятностью, что попаданий будет меньше двух (то есть либо 0, либо 1 попадание). Преобразование будет следующим:

```python
# Given values
n = 8        # total shots
p = 0.3      # probability of hitting the target
q = 1 - p    # probability of missing

# Calculate P(X = 0) and P(X = 1) using the binomial formula
P_X_0 = math.comb(n, 0) * (p ** 0) * (q ** n)
P_X_1 = math.comb(n, 1) * (p ** 1) * (q ** (n - 1))

# Probability of hitting the target at least twice
P_X_ge_2 = 1 - (P_X_0 + P_X_1)

# Round the result to three decimal places
round(P_X_ge_2, 3)
```

Задание 4

В магазине 7 покупателей. Каждый может совершить покупку с вероятностью 0,4. Найти вероятность того, что не более двух человек совершат покупку. Ответ округлите до сотых. В качестве десятичного разделителя используйте точку.

```python
import math

# Given values
n = 7       # total customers
p = 0.4     # probability of making a purchase
q = 1 - p   # probability of not making a purchase

# Calculate P(X = 0), P(X = 1), and P(X = 2)
P_X_0 = math.comb(n, 0) * (p ** 0) * (q ** n)
P_X_1 = math.comb(n, 1) * (p ** 1) * (q ** (n - 1))
P_X_2 = math.comb(n, 2) * (p ** 2) * (q ** (n - 2))

# Probability of at most two customers making a purchase
P_X_le_2 = P_X_0 + P_X_1 + P_X_2

# Round the result to two decimal places
round(P_X_le_2, 2)
```

Задание 5

В телеателье имеется 7 телевизоров. Для каждого телевизора вероятность того, что в данный момент он включен, равна 0,5. Найти вероятность того, что в данный момент включены не менее трех телевизоров. Ответ округлите до тысячных. В качестве десятичного разделителя используйте точку.

```python
import math

# Дано
n = 7        # общее количество телевизоров
p = 0.5      # вероятность того, что телевизор включен
q = 1 - p    # вероятность того, что телевизор выключен

# Вычисляем вероятность того, что включены ровно 0, 1 и 2 телевизора
P_X_0 = math.comb(n, 0) * (p ** 0) * (q ** n)   # Вероятность, что включено 0 телевизоров
P_X_1 = math.comb(n, 1) * (p ** 1) * (q ** (n - 1))   # Вероятность, что включен 1 телевизор
P_X_2 = math.comb(n, 2) * (p ** 2) * (q ** (n - 2))   # Вероятность, что включены 2 телевизора

# Находим вероятность того, что включены 3 или более телевизоров
P_X_ge_3 = 1 - (P_X_0 + P_X_1 + P_X_2)

# Округляем результат до трех знаков после запятой
round(P_X_ge_3, 3)
```

Задание 6

Пусть вероятность того, что покупателю необходим компьютер с операционной системой Linux, равна 0.25. Найдите вероятность того, что из шести покупателей по крайней мере двум необходим компьютер с операционной системой Linux.

Ответ округлите до сотых, в качестве разделителя используйте точку.

```python
import math

# Дано
n = 6        # общее количество покупателей
p = 0.25     # вероятность того, что покупателю нужен компьютер с Linux
q = 1 - p    # вероятность того, что покупателю не нужен компьютер с Linux

# Вычисляем вероятность того, что 0 покупателей нуждаются в Linux
P_X_0 = math.comb(n, 0) * (p ** 0) * (q ** n)

# Вычисляем вероятность того, что 1 покупатель нуждается в Linux
P_X_1 = math.comb(n, 1) * (p ** 1) * (q ** (n - 1))

# Находим вероятность того, что как минимум 2 покупателя нуждаются в Linux
P_X_ge_2 = 1 - (P_X_0 + P_X_1)

# Округляем результат до двух знаков после запятой
result = round(P_X_ge_2, 2)

# Выводим результат
print(result)
```

Задание 7

Для нормальной работы отдела по дата-аналитике необходимо, чтобы в офисе было не менее восьми программистов, а всего в штате 10 программистов. Вероятность того, что программист заболеет, равна 0.1. Найдите вероятность нормальной работы отдела по дата-аналитике.

Ответ округлите до сотых, а в качестве разделителя используйте точку.

```python
# Дано
n = 10      # общее количество программистов
p = 0.1     # вероятность того, что программист заболеет
q = 1 - p   # вероятность того, что программист не заболеет

# Вычисляем вероятности для 0, 1 и 2 заболевших программистов
P_X_0 = math.comb(n, 0) * (p ** 0) * (q ** n)      # Вероятность, что 0 программистов заболели
P_X_1 = math.comb(n, 1) * (p ** 1) * (q ** (n - 1)) # Вероятность, что 1 программист заболел
P_X_2 = math.comb(n, 2) * (p ** 2) * (q ** (n - 2)) # Вероятность, что 2 программиста заболели

# Находим общую вероятность нормальной работы отдела
P_X_le_2 = P_X_0 + P_X_1 + P_X_2

# Округляем результат до двух знаков после запятой
result = round(P_X_le_2, 2)

# Выводим результат
result
```

Задание 8

Среди всех скриптов, которые пишет программист, с 10%-ой вероятностью скрипт будет с ошибкой. Найдем вероятность того, что среди 30 скриптов два не будут работать.

Ответ округлите до тысячных, в качестве десятичного разделителя используйте точку.

```python
# Дано
n = 30      # общее количество скриптов
p = 0.1     # вероятность того, что скрипт с ошибкой
q = 1 - p   # вероятность того, что скрипт работает

k = 2       # количество скриптов с ошибкой

# Вычисляем вероятность того, что ровно 2 скрипта с ошибкой
P_X_2 = math.comb(n, k) * (p ** k) * (q ** (n - k))

# Округляем результат до трех знаков после запятой
result = round(P_X_2, 3)

# Выводим результат
result
```

Задание 9

В очень большом чате по дата аналитике сидит 1730 человек. Вероятность того, что день рождения наугад взятого участника чата приходится на определенный день года, равна 1/365 для каждого из 365 дней. Найдем наиболее вероятное число счастливчиков, которые родились 1 марта.

Для решения этой задачи вам может потребоваться эта формула.(биномиальное распределение)

```python
# Дано
n = 1730         # общее количество участников
p = 1 / 365      # вероятность рождения на 1 марта
q = 1 - p        # вероятность, что день рождения не на 1 марта

# Вычисляем np
np = n * p

# Вычисляем границы для наивероятнейшего числа
k_min = np - q  # нижняя граница
k_max = np + p  # верхняя граница

# Находим целые значения, которые могут быть наиболее вероятными
k_min_int = (k_min)  # округляем вниз
k_max_int = (k_max)  # округляем вниз

# Результаты
print("Границы:", (k_min_int, k_max_int))
```

Задание 10

В урне 20 шаров: 15 белых и 5 чёрных. Вынули подряд 5 шаров, причём каждый вынутый шар возвращается в урну и перед извлечением следующего шары в урне тщательно перемешиваются. Найдите вероятность того, что из пяти вынутых шаров будет 2 белых.

Ответ округлите до тысячных, в качестве десятичного разделителя используйте точку.

```python
import math

# Дано
n = 5            # общее количество вынутых шаров
k = 2            # количество белых шаров
p = 15 / 20      # вероятность вытянуть белый шар
q = 5 / 20       # вероятность вытянуть черный шар

# Биномиальный коэффициент
C_n_k = math.comb(n, k)

# Вероятность получить k белых шаров
probability = C_n_k * (p ** k) * (q ** (n - k))

# Округляем до тысячных
probability_rounded = round(probability, 3)

# Результат
print("Вероятность того, что из пяти вынутых шаров будет 2 белых:", probability_rounded)
```
