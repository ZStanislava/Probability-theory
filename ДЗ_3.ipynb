Задание 1

Из 500 компьютеров  180 принадлежат к 1 партии, 170 – ко второй партии, остальные к третьей. В первой партии 3% брака, во второй — 2%, в третьей – 6%. Случайно выбирается один компьютер. Определить вероятность того, что выбранный компьютер – бракованный.  Округлите полученный ответ до десятитысячных. В качестве разделителя используйте точку.

```python
# Данные задачи
P_A1 = 180 / 500
P_A2 = 170 / 500
P_A3 = 150 / 500

P_B_given_A1 = 0.03
P_B_given_A2 = 0.02
P_B_given_A3 = 0.06

# Вычисление полной вероятности события B
P_B = (P_B_given_A1 * P_A1) + (P_B_given_A2 * P_A2) + (P_B_given_A3 * P_A3)
round(P_B, 4)
```

Вероятность сделать работу за один день для Java-программиста — 0,85, для программиста на языке C++ 0,9, а для Python программиста 0,8. Найти вероятность того, что программист, выбранный наугад, сделает работу за один день. Ответ округлите до сотых. В качестве десятичного разделителя используйте точку.

Задание 3

Из 40 снайперов 18 попадает в цель с вероятностью 0,9, 8 — с вероятностью 0,4 и 14 – с вероятностью 0,7. Случайно выбранный снайпер произвел выстрел, поразив цель. Мы хотим вычислить, снайпер из какой группы наиболее вероятно совершил этот выстрел. 

В ответ запишите наибольшую вероятность, округленную до тысячных. В качестве разделителя используйте точку.

```python
# Данные задачи
P_A1 = 18 / 40
P_A2 = 8 / 40
P_A3 = 14 / 40

P_B_given_A1 = 0.9
P_B_given_A2 = 0.4
P_B_given_A3 = 0.7

# Полная вероятность события B
P_B = (P_B_given_A1 * P_A1) + (P_B_given_A2 * P_A2) + (P_B_given_A3 * P_A3)

# Апостериорные вероятности по формуле Байеса
P_A1_given_B = (P_B_given_A1 * P_A1) / P_B
P_A2_given_B = (P_B_given_A2 * P_A2) / P_B
P_A3_given_B = (P_B_given_A3 * P_A3) / P_B

# Найдем наибольшую апостериорную вероятность и округлим до тысячных
max_prob = round(max(P_A1_given_B, P_A2_given_B, P_A3_given_B), 3)
max_prob
```

Задание 4

Есть 4 кубика. На трех из них окрашена белым половина граней, а на четвертом кубике всего одна грань из шести белая. Наудачу выбранный кубик подбрасывается семь раз. Найти вероятность того, что был выбран четвертый кубик, если при семи подбрасываниях белая грань выпала ровно один раз.  Ответ округлите до тысячных. Используйте точку в качестве десятичного разделителя

```python
from math import comb

# Априорные вероятности для каждого кубика
P_A1 = P_A2 = P_A3 = P_A4 = 1 / 4

# Условные вероятности P(B | A_i) по формуле биномиального распределения
# Для первых трех кубиков (с вероятностью белой грани 1/2)
P_B_given_A1 = comb(7, 1) * (1/2)**1 * (1/2)**6
# Для четвертого кубика (с вероятностью белой грани 1/6)
P_B_given_A4 = comb(7, 1) * (1/6)**1 * (5/6)**6

# Полная вероятность события B
P_B = (P_B_given_A1 * P_A1) + (P_B_given_A1 * P_A2) + (P_B_given_A1 * P_A3) + (P_B_given_A4 * P_A4)

# Вероятность того, что выбран четвертый кубик, если белая грань выпала ровно один раз
P_A4_given_B = (P_B_given_A4 * P_A4) / P_B
round(P_A4_given_B, 3)
```

Задание 6

Известно, что около 60% всех людей в городе болеют заболеванием А. Для определения того, заболел человек или нет, проводится тест, точность которого составляет 99%. Вы получили положительный тест, какова вероятность того, что вы реально больны?

Округлите ответ до тысячных, в качестве десятичного раделителя используйте точку.

```python
# Данные задачи
P_A = 0.6       # Вероятность быть больным
P_not_A = 0.4   # Вероятность быть здоровым

P_T_positive_given_A = 0.99       # Вероятность положительного теста, если человек болен
P_T_positive_given_not_A = 0.01   # Вероятность положительного теста, если человек здоров

# Полная вероятность положительного теста
P_T_positive = (P_T_positive_given_A * P_A) + (P_T_positive_given_not_A * P_not_A)

# Вероятность того, что человек болен, если тест положительный
P_A_given_T_positive = (P_T_positive_given_A * P_A) / P_T_positive

# Округлим результат до тысячных
round(P_A_given_T_positive, 3)
```

Задание 7

Три фабрики производят лампочки. Фабрика А производит 20% всех лампочек, 50% лампочек производится на фабрике В и оставшиеся 30% на фабрике С. 2% лампочек, произведенных на фабрике А, 1% лампочек, произведенных на фабрике В, и 3% лампочек, произведенных на фабрике С, являются бракованными.На рынке наугад выбирается лампочка и оказывается бракованной. Какова вероятность того, что эта лампочка была произведена на заводе B?

Ответ округлите до тысячных, в качестве разделителя  используйте точку.

```python
# Данные задачи
P_A = 0.2       # Вероятность, что лампочка с фабрики A
P_B = 0.5       # Вероятность, что лампочка с фабрики B
P_C = 0.3       # Вероятность, что лампочка с фабрики C

P_D_given_A = 0.02  # Вероятность брака на фабрике A
P_D_given_B = 0.01  # Вероятность брака на фабрике B
P_D_given_C = 0.03  # Вероятность брака на фабрике C

# Полная вероятность брака
P_D = (P_D_given_A * P_A) + (P_D_given_B * P_B) + (P_D_given_C * P_C)

# Вероятность того, что бракованная лампочка с фабрики B
P_B_given_D = (P_D_given_B * P_B) / P_D

# Округлим результат до тысячных
round(P_B_given_D, 3)
```

Задание 8

Система РЛС (радиолокационная система) может обнаружить присутствие самолета в 98% случаев. Однако если в зоне действия радара нет самолета, система все равно сообщает (ложно), что самолет присутствует, в 5% случаев. В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что на радаре нет самолета?

Ответ округлите до десятитысячных, в качестве десятичного разделителя используйте точку.

```python

# Данные задачи
P_S = 0.07  # Вероятность, что самолет в зоне радара
P_not_S = 0.93  # Вероятность, что самолета нет в зоне радара

P_no_signal_given_S = 0.02  # Вероятность, что радар не зафиксирует самолет, если он есть
P_no_signal_given_not_S = 0.95  # Вероятность, что радар не зафиксирует самолет, если его нет

# Полная вероятность отсутствия сигнала
P_no_signal = (P_no_signal_given_S * P_S) + (P_no_signal_given_not_S * P_not_S)

# Вероятность того, что самолет находится в зоне радара, учитывая отсутствие сигнала
P_S_given_no_signal = (P_no_signal_given_S * P_S) / P_no_signal

# Округлим результат до десятитысячных
result = round(P_S_given_no_signal, 4)

print(result)
```

Задание 9

Система РЛС (радиолокационная система) может обнаружить присутствие самолета в 98% случаев. Однако если в зоне действия радара нет самолета, система все равно сообщает (ложно), что самолет присутствует, в 5% случаев. В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что радар показывает наличие самолета?

Ответ округлите до тысячных, в качестве десятичного разделителя используйте точку.

```python
# Заданные вероятности
P_S = 0.07                      # Вероятность, что самолет в зоне радара
P_not_S = 1 - P_S               # Вероятность, что самолета нет в зоне радара
P_signal_given_S = 0.98         # Вероятность обнаружения самолета, если он есть
P_signal_given_not_S = 0.05     # Вероятность ложного сигнала, если самолета нет

# Полная вероятность сигнала
P_signal = (P_signal_given_S * P_S) + (P_signal_given_not_S * P_not_S)

# Применяем теорему Байеса для нахождения P(S | сигнал)
P_S_given_signal = (P_signal_given_S * P_S) / P_signal

# Округлим результат до тысячных
round(P_S_given_signal, 3)
```

Задание 10

Передатчик аварийного локатора самолета (ELT) - это устройство, предназначенное для передачи сигнала в случае аварии. Компания A производит 80% всех ELT на рынке, компания B - 15%, а компания C - остальные 5%. Уровень дефектности ELT, произведенных компанией A, составляет 4%, компанией B - 6%, а компанией C - 9%. Случайно выбранный ELT был протестирован и оказался бракованным. Найдите вероятность того, что он был произведен компанией A.

Округлите ответ до тысячных, в качестве десятичного разделителя используйте точку.

```python
# Данные задачи
P_A = 0.80  # Вероятность, что ELT произведен компанией A
P_B = 0.15  # Вероятность, что ELT произведен компанией B
P_C = 0.05  # Вероятность, что ELT произведен компанией C

P_D_given_A = 0.04  # Вероятность бракованного ELT от компании A
P_D_given_B = 0.06  # Вероятность бракованного ELT от компании B
P_D_given_C = 0.09  # Вероятность бракованного ELT от компании C

# Полная вероятность того, что ELT бракованный
P_D = (P_D_given_A * P_A) + (P_D_given_B * P_B) + (P_D_given_C * P_C)

# Вероятность того, что бракованный ELT был произведен компанией A
P_A_given_D = (P_D_given_A * P_A) / P_D

# Округляем результат до тысячных
result = round(P_A_given_D, 3)
print(result)
```
